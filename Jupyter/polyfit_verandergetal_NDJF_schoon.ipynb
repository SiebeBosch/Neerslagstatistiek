NDJF --> Tweedegraads polynoom: 5.952e-06 * D^2 - 1.515e-03 * D + 1.277

```python
def poly(D):
    return 5.952e-06 * D**2 - 1.515e-03 * D + 1.277
```

```python
poly(36)
```

```python
def verandergetalfunctie(Ts,D,T=1):
    # Ts: temperatuurstijging in graden Celsius
    # D:  duur in uren
    # T:  terugkeertijd (irrelevante parameter)

    if D < 1/6: 
        raise ValueError(f'Gekozen duur {D} valt buiten domein (10 minuten t/m 240 uur)')
    elif D <= 24:
        v = 1.244
    elif D < 120:
        v = poly(D)
    elif D <= 240:
        v = 1.181
    elif D > 240:
        raise ValueError(f'Gekozen duur {D} valt buiten domein: 10 minuten t/m 10 dagen (240 uur)')
  
    return 1 + (v-1) * (Ts-0.4)/(4-0.4) 
    # de factor v is afgeleid voor 4 graden temperatuurstijging t.o.v. 2005, 
    # maar in 2023 hebben we al 0.4 graden gehad (0.6 graden in 2033, ~0.4 in 2023)
```

```python
def locatieparameter_KNMI_2019_NDJF(Duur):
    # Invoer: Duur in minuten
    # Invoer:Herhalingstijd in jaar 

    if Duur <= 720 : # Duur is kleiner dan 12u
        
        # Eq. (12) pagina 20 (33 van pdf van deelrapport 1 STOWA2019-19)
        
        loc = 1.07*1.02*(4.883 - 5.587 * np.log10(Duur) + 3.526 * (np.log10(Duur))**2)
    
    else: # Duur is groter dan 12u
        
        # Eq. (15) pagina 21 (34 van pdf van deelrapport 1 STOWA2019-19)
        # Formule gebruikt de de duur in uren; deel de duur in minuten door 60 (!)
        
        loc = (0.670 - 0.0426 * np.log(Duur/60))**(-1.0/0.193)
    
    return loc

def dispersiecoefficient_KNMI_2019_NDJF(Duur): 
    # Invoer: Duur in minuten
    # Invoer:Herhalingstijd in jaar 

    if Duur <= 720 : # Duur is kleiner dan 12u
        
        # Eq. (13) pagina 20 (33 van pdf van deelrapport 1 STOWA2019-19)
        
        if Duur > 91:            
            
            disp = 0.2684

        elif Duur <= 91:
                    
            disp = 0.41692 - 0.07583 * np.log10(Duur)

            
    else: # Duur is groter dan 12u
        
        # Eq. (16) pagina 21 (34 van pdf van deelrapport 1 STOWA2019-19)    
        
        disp = 0.234
        
    return disp

def vormparameter_KNMI_2019_NDJF(Duur):
    # Invoer: Duur in minuten (!)
    if Duur <= 720 : # Duur is kleiner dan 12u

        # Eq. (14) pagina 20 (33 van pdf van deelrapport 1 STOWA2019-19) 
                    
        vorm = -0.294 + 0.1474 * np.log10(Duur) - 0.0192 * (np.log10(Duur))**2
                   
    else: # Duur is groter dan 12u

        # Eq. (16) pagina 21 (34 van pdf van deelrapport 1 STOWA2019-19)        
        # formule gebruikt de duur in uren, daarom de duur in minuten door 60 delen (!)
        
        vorm = -0.090 + 0.017 * Duur/(60*24)
    
    return vorm

def vol_KNMI_2019_NDJF(Duur, T):
    # Invoer: Duur in minuten (!)
    # Invoer: Herhalingstijd in jaar 
    # Uitvoer: Volume in mm

    Locpar = locatieparameter_KNMI_2019_NDJF(Duur)
    Vormpar = vormparameter_KNMI_2019_NDJF(Duur)
    dispcoeff = dispersiecoefficient_KNMI_2019_NDJF(Duur)
    Schaalpar = dispcoeff * Locpar
    
    # Alleen GEV-verdeling voor NDJF-statistiek
    Vol_KNMI_2019 = Locpar + (Schaalpar / Vormpar) * (1- (1/T)**Vormpar)
            
    return Vol_KNMI_2019
```

```python
Ts = [0.5, 1, 2, 5, 10, 20, 25, 50, 100, 200, 250, 500, 1000] # in jaren
Ts_str = ['0.5 jaar', '1 jaar', '2 jaar', '5 jaar', '10 jaar', '20 jaar', '25 jaar', '50 jaar', '100 jaar', '200 jaar', '250 jaar', '500 jaar', '1000 jaar'] # in jaren
dduren = [1/6, 1/2, 1, 2, 4, 8, 12, 24, 48, 72, 96, 192, 240] # in uren, 240 uur = 10 dagen
dduren_str = ['10 min', '30 min', '60 min', '2 uur', '4 uur', '8 uur', '12 uur', '24 uur', '2 dagen', '3 dagen', '4 dagen', '8 dagen', '10 dagen'] 

scenarios = dict({'huidig': 0.4, '2033L': 0.6, '2100Ln': 0.8,'2050Mn': 1.1,'2100Mn': 1.9,'2150Mn': 2.1,'2050Hn': 1.5,'2100Hn': 4,'2150Hn': 5.5, 
                  '2100Ld': 0.8, '2050Md': 1.1,'2100Md': 1.9,'2150Md': 2.1,'2050Hd': 1.5,'2100Hd': 4,'2150Hd': 5.5, '2005': 0.0})

for k, s in enumerate(scenarios):
    temp = scenarios[s]
    df_vg_NDJF = pd.DataFrame(index=Ts_str, columns=dduren_str)
    df_vg_NDJF2 = pd.DataFrame(index=Ts_str, columns=dduren_str)
    df_vol_NDJF = pd.DataFrame(index=Ts_str, columns=dduren_str)
    
    for i, T in enumerate(Ts):
        for j, duur in enumerate(dduren):
            df_vg_NDJF.iat[i,j] = verandergetalfunctie(temp,duur,T)
            df_vg_NDJF2.iat[i,j] = (verandergetalfunctie(temp,duur,T)-1)*100
            df_vol_NDJF.iat[i,j] = verandergetalfunctie(temp,duur,T) * vol_KNMI_2019_NDJF(60*duur, T)
    
    # Write to sheets in Excel workbook
    if k == 0:
        with pd.ExcelWriter(f"Statistiek_NDJF.xlsx") as writer:
            df_vol_NDJF.to_excel(writer, sheet_name=f"{s}", float_format="%.1f")    
    else:
        with pd.ExcelWriter(f"Statistiek_NDJF.xlsx", mode = 'a') as writer:
            df_vol_NDJF.to_excel(writer, sheet_name=f"{s}", float_format="%.1f")
```

```python
duren = [1, 2, 4, 8, 12, 24, 48, 96, 192, 240]
Ts = [2, 1, 1/2, 1/5, 1/10, 1/20, 1/25, 1/50, 1/100, 1/120] #, 1/130, 1/140, 1/150, 1/160, 1/165, 1/200, 1/250, 1/500, 1/1000]

plt.close('all')
mpl.rcParams["axes.formatter.useoffset"] = False
font_axislabels = font_manager.FontProperties(family='Verdana',
                                   weight='bold',
                                   style='normal', size=14)
font_legend_axisticks = font_manager.FontProperties(family='Verdana',
                                   style='normal', size=14)
f , axarr = plt.subplots(5, 2, figsize=(10,20), dpi=200)
legend_handles = []
nr_legend_handles = 2
n_leg_cols = 1    

for i, duur in enumerate(duren):
    # huidige fit
    df_sortedMax_statline = pd.DataFrame({'Overschrijdingskans': Ts})
    df_sortedMax_statline['Stat2019Vol'] = df_sortedMax_statline['Overschrijdingskans'].apply(lambda x: vol_KNMI_2019_NDJF( 60*duur, 1/x))
    h, = axarr[i//2, i%2].plot(1/df_sortedMax_statline['Overschrijdingskans'],
                               df_sortedMax_statline['Stat2019Vol'], 
                               color = 'tab:gray', linewidth = 2, label = 'huidig klimaat - NDJF')
    legend_handles.append(h)

    df_vol_NDJF = pd.DataFrame({'Overschrijdingskans': Ts})
    df_vol_NDJF['Vol'] = df_sortedMax_statline['Stat2019Vol'] * df_vol_NDJF['Overschrijdingskans'].apply(lambda x: verandergetalfunctie(4.0,duur,1/x))
    h, = axarr[i//2, i%2].plot(1/df_sortedMax_statline['Overschrijdingskans'], df_vol_NDJF['Vol'], 
                               color = 'tab:red', linestyle = 'dashed', linewidth = 2, label = '2100Hn/2100Hd - NDJF')
    legend_handles.append(h)    
    
    axarr[i//2, i%2].set_xscale('log')
    axarr[i//2, i%2].xaxis.set_major_formatter( matplotlib.ticker.ScalarFormatter())
    axarr[i//2, i%2].xaxis.get_major_formatter().set_scientific(False)

    axarr[i//2, i%2].set_title('Duur: ' + str(duur) + ' uur', fontproperties=font_legend_axisticks)

    if i//2 == 4:
        axarr[i//2, i%2].set_xlabel('Terugkeertijd (jaar)', fontproperties=font_axislabels, labelpad=15)

    if i%2 == 0:
        axarr[i//2, i%2].set_ylabel('Neerslagvolume (mm)', fontproperties=font_axislabels, labelpad=15)

for i in range(len(duren)):
    axarr[i//2, i%2].set_ylim([20*math.floor(axarr[i//2, 0].get_ylim()[0]/20),  
                               20*math.ceil(axarr[i//2, 1].get_ylim()[1]/20)])
    axarr[i//2, i%2].set_yticks(np.arange(axarr[i//2, i%2].get_ylim()[0],
                                          axarr[i//2, i%2].get_ylim()[1]+20, step=20))
    axarr[i//2, i%2].set_xlim(axarr[0, i%2].get_xlim())            

plt.setp([a.get_yticklabels() for a in axarr[:, 0]], fontproperties = font_legend_axisticks)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False)

f.legend(legend_handles, [i.get_label() for i in legend_handles[0:nr_legend_handles]],  
         loc = 'upper center', ncol=n_leg_cols, prop = font_legend_axisticks)
f.align_ylabels(axarr[:, 0])

plt.savefig("Neerslagvolumina_duren_NDJF")
```
